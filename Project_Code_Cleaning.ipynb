```python
import pandas as pd
```

```python
#importing the csv file
df = pd.read_csv('raw_sales.csv')
```

```python
df
```

```python
df.dtypes
```

```python
#Channging the date column to date format
df['datesold'] = pd.to_datetime(df['datesold'])
```

```python
df
```

```python
df.dtypes
```

```python
#Extracting the months from the date column
df['month'] = df['datesold'].dt.month
```

```python
#Extracting the year from date column
df['year'] = df['datesold'].dt.year
```

```python
df
```

```python
df.sort_values(['year','month'])
```

```python
#Creating a month-year column for cleaning
df['year_month'] = df['datesold'].dt.strftime('%Y-%m')
```

```python
df
```

```python
#Sorting the values based on year-month column
df = df.sort_values('year_month')
```

```python
df
```

```python
#Grouping the price by month and year and taking mean of the price
avg_price_by_month = df.groupby(["month","year"]).agg({"price": "mean"})
```

```python
avg_price_by_month
```

```python
avg_price_by_month = avg_price_by_month.reset_index()
```

```python
avg_price_by_month
```

```python
#Sorting the values by year and then month
avg_price_by_month=avg_price_by_month.sort_values(['year','month'])
```

```python
avg_price_by_month
```

```python
#Final dataset used for the analysis
avg_price_by_month.to_excel('cleaned_data.xlsx')
```
